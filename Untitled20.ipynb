```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen2.5-3B-Instruct-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
)
```

```python
# 1. Define the prompt style
prompt_style = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Response:
<thought>
{}
</thought>
{}"""

# 2. Update the formatting function
def format_prompts(examples):
    questions = examples["Question"]
    thoughts  = examples["Complex_CoT"]
    answers   = examples["Response"]
    texts = []
    for q, t, a in zip(questions, thoughts, answers):
        text = prompt_style.format(q, t, a) + tokenizer.eos_token
        texts.append(text)
    return { "text" : texts, }

# 3. Re-load and map the dataset, ensuring we remove original columns
from datasets import load_dataset
dataset = load_dataset("FreedomIntelligence/medical-o1-reasoning-SFT", 'en', split = "train[:2000]")

# We must remove the original columns so the Trainer doesn't try to tokenize them
column_names = dataset.column_names
dataset = dataset.map(format_prompts, batched = True, remove_columns = column_names)

print("✅ Dataset mapped successfully. Remaining columns:", dataset.column_names)
```

```python
from trl import SFTConfig, SFTTrainer
from transformers import TrainingArguments
import torch
import gc

# 1. Clear GPU memory completely
if 'trainer' in locals(): del trainer
gc.collect()
torch.cuda.empty_cache()

# 2. Use SFTConfig with packing=True to resolve the ValueError
args = SFTConfig(
    per_device_train_batch_size = 1,
    gradient_accumulation_steps = 8,
    warmup_steps = 10,
    max_steps = 60,
    learning_rate = 2e-4,
    fp16 = not torch.cuda.is_bf16_supported(),
    bf16 = torch.cuda.is_bf16_supported(),
    logging_steps = 1,
    optim = "adamw_8bit",
    weight_decay = 0.01,
    lr_scheduler_type = "linear",
    seed = 3407,
    output_dir = "outputs",
    gradient_checkpointing = True,
    max_seq_length = 1024,
    packing = True, # Changed to True to fix tensor creation error
    dataset_text_field = "text",
)

# 3. Initialize the Trainer
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    args = args,
)

# 4. Start Training
trainer.train()
```

```python
FastLanguageModel.for_inference(model)
case = """A 32-year-old female presents with sudden onset sharp chest pain and shortness of breath.
She recently arrived on a 14-hour international flight. Her only medication is an oral contraceptive pill.
On exam, she is tachycardic (HR 115) and her right calf is tender and significantly more swollen than the left.
What is the most likely diagnosis and the next immediate step?"""

inputs = tokenizer([prompt_style.format(case, "", "")], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 1024)
```
